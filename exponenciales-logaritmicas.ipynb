<h2>Funciones exponenciales y logarítmicas</h2>

Otro tipo de funciones trascendentes son las funciones exponenciales de la forma $f(x)=a^x$ y sus inversas denominadas logarítmicas.

<h2>Funciones exponenciales elementales</h2>

La forma más simple de las funciones exponenciales es $f(x)=a^x=e^{bx}$, donde $a\, y\, b\, \in  \mathbb{R}$ con $a>0\land a\neq 1$, $b=\ln{a}$. El caso de $a\neq 1$ se excluye porque $1^x$ simplemente es igual a 1 para todo $x$, mientras que las bases negativas quedan excluidas también debido a que si bien para x enteros la fórmula está bien definida para el caso de valores fraccionales deberíamos considerar el campo del análisis de los números complejos (ver breve nota al final del cuaderno).


Como se observa, la función consiste en elevar un número real ( con las restrincciones mencionadas) a algun otro número real. 

Conocemos el valor de $f(x)=a^x$ cuando $x$ es un número natural, un número entero o una fracción, qué occurrirá con un número iracional como $\sqrt{2}$?

La expresión decimal infinita $\sqrt{2} \sim 1.4142214 ...$ es simplemente otra manera de decir que $\sqrt{2}$ es un número real $r$ que satisface las siguientes desigualdades:

$1< r < 2$

$1.4< r < 1.5$

$1.41< r < 1.42$

$1.414< r < 1.415$

$...$

Entonces existe un único número real $s$ que satisface:

$a^{1}< s < a^{2}$

$a^{1.4}< s < a^{1.5}$

$a^{1.41}< s < a^{1.42}$

$a^{1.414}< s <a^{1.415}$

$...$

Luego resulta razonable definir $a^{\sqrt{2}}$ como ese único número $s$. 

Este proceso se puede realizar para cualquier número irracional y, por lo tanto tenemos definida la función $f(x)=a^x$ para todo número real.

$\textbf{Ejemplo 1.}$ Gráfica de una curva exponencial (para caso en que la base es $a=e$ obtenemos la curva exponencial natural $y=e^x$) 

Notar: 

* $f(x)=a^{-x}=(\frac{1}{a})^x=\frac{1}{a^x}$

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5.0, 5.0)
bases = [2,np.e,10]
colors = ['red','blue','black']
fn = lambda base,x: base**x
#xData = array([0.15,2.3,3.15,4.85,6.25,7.95])
#yData = array()
#yData = array([fn(xData[i]) for i in range(len(xData))])
fig, ax = plt.subplots(1, 2,figsize=(10, 5))
for i in range(len(bases)):
    ax[0].plot(x, [fn(bases[i],x[j]) for j in range(len(x))], colors[i], lw=3, label="$f(x)=%.5s^x$"%bases[i]) #row=0, col=0

for i in range(len(bases)):
    ax[1].plot(x, [fn(1/(bases[i]*1.),x[j]) for j in range(len(x))], colors[i], lw=3, label="$f(x)=%.5s^{-x}$"%bases[i]) #row=0, col=0

    #ax[1].plot(x, y2, 'g',lw=3,label="$f(x)=2^{-x}$") #row=0, col=1
#Ejes horizontal/vertical
ax[0].axhline(0, color='black')
ax[0].axvline(0, color='black')
ax[1].axhline(0, color='black')
ax[1].axvline(0, color='black')
#Nombre ejes
ax[1].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].set_ylim(0,10)
ax[1].set_ylim(0,10)

ax[0].legend(loc='upper left', shadow=True)
ax[1].legend(loc='upper right', shadow=True)

plt.show()
```

<h3> Observaciones:</h3>

* La función solo toma valores positivos, por lo que $Dom(f)= \mathbb{R}$ y la $Im(f)= \mathbb{R}^+$

* Para $a>1$ ($b>0$, $b=\ln{a}$) es monótona creciente desde 0 hasta $\infty$

* Para $a<1$ ($b<0$) es monótona decreciente desde $\infty$ hasta 0, más rápidamente (discutir brevemente concepto de la pendiente de la curva en un punto) cuanto mayor sea $|b|$

* La curva se aproxima asintóticamente al eje x más rápidamente cuanto mayor sea $|b|$, para b>0 lo hace por la izquierda y para b<0 por la derecha.

* f(0)=1 para cualquier valor de la base, puede explicar por qué?

<h2>Funciones logarítmicas elementales</h2>

Como la función $f(x)=a^x$ es creciente o decreciente en su dominio, es inyectiva en el mismo y, por lo tanto admite función inversa. La función inversa se llama función logarítmica en base a y se simboliza $g(x)=\log_a{x}$.

Recordar que, por tratarse de funciones inversas se cumple:

$y=\log_a{x}$ sii $a^y=x$ para $x\in \mathbb{R}^+$ e $y\in \mathbb{R}$

$\textbf{Ejemplo 2.}$ Gráfica de curvas logarítmicas

Notar que las curvas logarítmicas se obtienen de la reflexión de la curva exponencial con respecto a la bisectriz $y=x$

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import math

x = np.linspace(1e-8, 5.0,100)
bases = [2,np.e,10]
colors = ['red','blue','black']
fn = lambda base,x: math.log(x,base)
#xData = array([0.15,2.3,3.15,4.85,6.25,7.95])
#yData = array()
#yData = array([fn(xData[i]) for i in range(len(xData))])
fig, ax = plt.subplots(1, 2,figsize=(10, 5))
for i in range(len(bases)):
    ax[0].plot(x, [fn(bases[i],x[j]) for j in range(len(x))], colors[i], lw=3, label="$g(x)=log_{%.5s}(x)$"%bases[i]) #row=0, col=0

for i in range(len(bases)):
    ax[1].plot(x, [fn(1/(bases[i]*1.),x[j]) for j in range(len(x))], colors[i], lw=3, label="$g(x)=log_{1/%.5s}(x)$"%bases[i]) #row=0, col=0

    #ax[1].plot(x, y2, 'g',lw=3,label="$f(x)=2^{-x}$") #row=0, col=1
#Ejes horizontal/vertical
ax[0].axhline(0, color='black')
ax[0].axvline(0, color='black')
ax[1].axhline(0, color='black')
ax[1].axvline(0, color='black')
#Nombre ejes
ax[1].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].set_xlim(0,5)
ax[0].set_ylim(-5,5)
ax[1].set_xlim(0,5)
ax[1].set_ylim(-5,5)

ax[0].legend(loc='upper left', shadow=True)
ax[1].legend(loc='upper right', shadow=True)

plt.show()
```

<h3> Observaciones:</h3>

* La función está definida para valores positivos de la variable independiente, por lo que $Dom(f)= \mathbb{R}^+$ y $Im(f)= \mathbb{R}$

* Para $a>1$ es monótona creciente desde $-\infty$ hasta $\infty$

* Para $a<1$ es monótona decreciente desde $\infty$ hasta $-\infty$, más lentamente tanto mayor sea $|\ln{a}|$

* La curva pasa por el punto (1,0) y asintóticamente se aproxima al eje y más rápidamente cuanto mayor sea $|\ln{a}|$, con a>1 hacia abajo y con a<1 hacia arriba (en el eje de las y's)

<h3> Propiedades</h3>

1) $\log_a{(xy)}=\log_a{x}+\log_a{y}$, pues $a^{x+y}=a^xa^y$

2) $\log_a{(x^n)}=n\log_a{x}$, pues $a^{nx}=(a^x)^n$

3) $\log_a{(x/y)}=\log_a{x}-\log_a{y}$, pues $a^{x-y}=a^x/a^y$

4) Cambio de base: $\log_a{x}=\frac{1}{\log_b{a}}\log_b{x}$ (Demostrar)


<h3> Por qué no estudiaremos el caso de bases $a<0$?</h3>

Para el caso $f(x)=(-a)^x$, con a>0 debemos considerar el campo del análisis de números complejos que se verá en cursos más avanzados. Sin embargo daremos aquí una breve introducción para motivar al lector a continuar su travesía en el estudio de uno de los tantos  apasionantes campos las matemáticas que ha tenido grandes y muy diversas aplicaciones en las ciencias y en las ingenierías.

Los números complejos $z$ se pueden definir como pares ordenados de números reales que escribimos como $z=(x,y)=x+iy$, donde $i^2=-1$. Podemos rescribir la función utilizando la **fórmula de Euler**: $e^{ix}=\cos(x)+i\sin(x)$, de la siguiente manera: 

$f(x)=(-a)^x=(a(-1)(1))^x=(a e^{i\pi} e^{i(2\pi k) })^x=(ae^{i\pi(1+2k)})^x=a^xe^{i\pi(1+2k)x}=a^x(\cos(\pi(1+2k)x)+i\sin(\pi(1+2k)x))$

, con $k \in \mathbb{Z}$ 

Esta es la forma que se utiliza cuando $a$ y $x$ son reales, para cada $x$ obtenemos múltiples números complejos: uno para cada valor de $k$. Cuando $x$ es racional obtendremos un número finito de valores para $f(x)$ debido al comportamiento periódico de la función, esto es, algunos valores de k darán argumentos de las funciones trigonométricas con valores en 2$\pi$ mas que el valor anterior a $k$.
