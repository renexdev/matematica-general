<h3>Introducción</h3>

En matemática, una función es una $\textbf{regla}$ que asigna a cada elemento de un conjunto de salida un elemento de otro conjunto de llegada (que puede coincidir con el primer conjunto).

Trataremos con funciones que a cada número real de un cierto conjunto prefijado, le corresponde otro número real. Tales funciones se denominan funciones de variable real y a valores reales.

Retomemos la definición de función [1]


 > $f$ es una función o aplicación de A=$\mathbb{R}$ en B=$\mathbb{R}$, $f$ : $\mathbb{R} \rightarrow \mathbb{R}$, sii $f$ es una relación entre A y B, tal que todo elemento de A tiene un único correspondiente en B.

o expresado matemáticamente:
 > $f$ es una función de A=$\mathbb{R}$ en $B=\mathbb{R}$ sii $f$ es un subconjunto de AxB que satisface:

> i) $\forall x \in\mathbb{R}$, $\exists y\in\mathbb{R}$/ $(x,y) \in\, f$ (Existencia)

> ii) $(x,y) \in\, f \land$ $(x,z) \in\, f \Rightarrow y=z$ (Unicidad)

El $\textbf{dominio}$ de la función $f$ es la totalidad de los elemento del conjunto de salida ($\mathbb{R}$) que admiten imagen en el conjunto de llegada ($\mathbb{R}$)

> $Dom(f)$ = {x $\in \mathbb{R} / (x,y) \in f(x)$}

Notar: cuando tenemos una fórmula para definir la función, el dominio se define como aquellos números reales para los cuales “podemos hacer la cuenta”. Algunas operaciones tienen sus restricciones:

>* Un denominador no puede ser cero

>* Una raíz cuadrada no puede calcularse sobre números negativos

>* Algunas funciones como el logaritmo no pueden calcularse sobre números negativos ni sobre cero (ya lo veremos en detalle más adelante)


Definimos la $\textbf{imagen}$ (codominio o rango) de una función $f$ como el conjunto de los elementos del conjunto de llegada ($\mathbb{R}$) que admiten un antecedente en el conjunto de salida($\mathbb{R}$)

> $Im(f)$ = {y $\in \mathbb{R} / \exists x \in  Dom(f), y=f(x)$}

Notar que una función queda especificada si se dan el dominio (o conjunto de salida), el codominio (conjunto de llegada) y además la relación f$\subset\mathbb{R}x\mathbb{R}$, que satisface las condiciones i) y ii) de la definición.

Las imágenes de las funciones anteriores estan dadas por:

<h2>Generalidades de las funciones</h2>

** Funciones implícitas y explícitas**

Las dos variables (dependiente e independiente) pueden venir dadas por la expresión $y=f(x)$, como se vio en los ejemplos: 1) $y=2x$, 2) $y=x^2$. Cuando está dada en esta forma se dice que está en la **forma explícita**. Sin embargo la podríamos haber expresado así: 1) $y-2x=0$, 2) $y-x^2=0$ 
, en general en la forma, $F(x,y)=0$. Cuando está dada en esta forma se dice que está en la forma implícita.

> No toda ecuación $F(x,y)=0$ define una función

Notar que en el caso del ejemplo **3)** escrito en forma implícita quedaría $y^2-x+1=0$, en este caso dejaría de ser una función ya que debemos considerar ambas ramas de la función raíz cuadrada (Recordar la definición de valor absoluto $|x|=\sqrt{x^2}$). 


**Raíces de una función** 

> Dada un función $f$ : $\mathbb{R} \rightarrow \mathbb{R}$ definida por f(x), los valores de $x$ para los cuales $f(x)=0$ se conocen como las raíces o ceros de la función.

Gráficamente los zeros de la función son los puntos donde la función intersecta el eje de las abscisas (eje de las $x$).

La determinación de las raíces reales de una función, si es que existen, nos permitirán definir los intervalos en los cuales la función es positiva y los intervalos en los cuales es negativa.

**Conjunto de positividad** 

> Llamaremos conjunto de positividad de una función al conjunto formado por aquellos valores de x para los cuales $f(x) > 0$.

Notar que el conjunto de positividad de una función están formados por intervalos abiertos y gráficamente son los puntos en el eje $x$ donde la curva está sobre el eje.

**Conjunto de negatividad**

> Llamaremos conjunto de negatividad de una función al conjunto formado por aquellos valores de x para los cuales $f(x) < 0$.

Notar que el conjunto de negatividad de una función están formados por intervalos abiertos y gráficamente son los puntos en el eje $x$ donde la curva está debajo del eje.

$\textbf{Ejemplo 6}$. Sea $f$ : $\mathbb{R} \rightarrow \mathbb{R}$ definida por $f(x)=x^2-4$. Determinar las raíces de la función y los conjuntos de positividad y negatividad. Graficar


```python
%matplotlib inline
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,5, 50)
def cuadratica(a,b,c,x):
    return a*x**2+b*x+c

a=1; b=0; c=-4
#genera el obj figure
plt.figure()
#Eje horizontal
plt.axhline(0, color='black')
#Eje vertical
plt.axvline(0, color='black')
#Grafico la funcion
plt.plot(x,cuadratica(a,b,c,x),'b-',label="$f(x)=x^2-4$")
#Elijo dos puntos
plt.scatter([1,4],[cuadratica(a,b,c,1),cuadratica(a,b,c,4)])
plt.text(1, cuadratica(a,b,c,1)+1,'(1 , %s)'%cuadratica(a,b,c,1),horizontalalignment='center', verticalalignment='center')
plt.text(4, cuadratica(a,b,c,4)+1,'(4 , %s)'%cuadratica(a,b,c,4),horizontalalignment='center', verticalalignment='center')
#Establezco limites para graficar
plt.xlim(-6, 6)
plt.ylim(-5, 16)
#Titulo y nombre a los ejes
plt.title("Ejemplo 6")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc='best')
plt.show()
```

**Crecimiento y decrecimiento de una función**

Sea $f$ : $\mathbb{R} \rightarrow \mathbb{R}$ :

> Una función se dice **creciente** en un intervalo si para todo par de puntos dentro del intervalo (perteneciente al dominio de la función) se cumple que si $a < b$  entonces $f(a) < f(b)$

> Una función se dice **decreciente** en un intervalo si para todo par de puntos dentro del intervalo (perteneciente al dominio de la función) se cumple que si $a < b$  entonces $f(a) > f(b)$

**Simetría de funciones **

Sea $f$ : $\mathbb{R} \rightarrow \mathbb{R}$ :

> Una función $f$ se dice **par**  sii para todo  x$\in Dom(f)$ se cumple que $f(-x)=f(x)$.

> Una función $f$ se dice **impar**  sii para todo  x$\in Dom(f)$ se cumple que $f(-x)=-f(x)$.

Notas:

En la definición está implícito que para que una función sea par o impar una de las condiciones que debe cumplir el $Dom(f)$ es que $\forall x \in Dom(f)$ tamién que todo $-x \in Dom(f)$. Es decir, para que una función sea par o impar es condición necesaria que su dominio sea un conjunto simétrico con respecto al origen.

La gráfica de una función par es simétrica respecto al eje de las ordenadas, mientras que la de una función impar es simétrica respecto al origen de coordenadas.

<h2>Funciones Algebraicas</h2>

Una función $y=f(x)$ es **algebraica** si, dada la variable independiente $x$, para obtener $f(x)$ deben realizarse sobre la misma un número finito de operaciones algebraicas. Estas operaciones incluyen la suma, resta, multiplicación, división, elevación a potencias enteras o racionales. Un ejemplo de una función algebraica es la función $f$ : $\mathbb{R} \rightarrow \mathbb{R}$ definida por: $f(x)=\frac{(x^2-3x+1)^2}{(5x^{1/2}+x^{1/3})^ {1/5}}$

Dentro de las funciones algebraicas merecen especial atención por su importancia las **funciones polinómicas**

>Una función **polinómica de grado $n$** es una función $P_n$ : $\mathbb{R} \rightarrow \mathbb{R}$ : de la forma:
            
>$P_n(x)=a_nx^n+a_{n-1}x^{n-1}+...+a_1x+a_0$

>donde $n$ es un entero no negativo y $a_n\neq0$

Los números $a_0, a_1, a_2, ..., a_n$ se denomninan **coeficientes** del polinomio. 

El número $a_0$ es el coeficiene constante o término constante.

El número $a_n$, el coeficiene que acompaña la potencia más alta, es el coeficiente dominante y el término $a_nx^n$ se denomina término dominante.

El $Dom(P_n) = \mathbb{R}$ cualquiera sea $n$.

Observaciones:

* Si la función polinómica es de grado 1, obtenemos la función lineal, que discutimos más arriba.

* Si la función polinómica es de grado 0 obtenemos una función denominada constante

* Los polinomios con un término se llaman monomios.

$\textbf{Ejemplo 9.}$ Dar una expresión analítica para la siguiente función y determinar el dominio: Un rectángulo tiene un área de 10cm$²$, expresar el perímetro en función de la lonigtud de uno de sus lados.

Sean $x$ e $y$ las longitudes de los lados del rectángulo.

$A$ = $xy$ (área del rectángulo)
 $xy=10$ (1)
 
 Como el área es distanta de cero, tenemos que $x \neq 0$ e $y\neq 0$.
 
 Por definición de de perímetro: $P(x,y)=2x+2y$ (2)
 
 por (1) y como $x\neq 0$ tenems que $y=10/x$ que reemplazando en (2) queda
 
 $P(x)=2(x+\frac{10}{x})=\frac{2x^2+20}{x}$ (3)
 
 El dominio de (3) es $Dom(P)=\{x/x>0\}$, por lo que $x$ no puede ser menor que cero pues x representa la longitud de un lado del rectángulo. 
 
 Notar que el $Dom(P)$ no depende solamente de la fórmula (3) sino también del significado que tiene la variable $x$. 
 

```python
%matplotlib inline
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1e-10,6, 50)
def perimetro(x):
    return (2*x**2+20)/x

#genera el obj figure
plt.figure()
#Eje horizontal
plt.axhline(0, color='black')
#Eje vertical
plt.axvline(0, color='black')
#Grafico la funcion
plt.plot(x,perimetro(x),'g-',lw=2,label="Perímetro del rectángulo")
#Establezco limites para graficar
plt.xlim(0., 6.)
plt.ylim(0., 50)
#Titulo y nombre a los ejes
plt.title("Ejemplo 9")
plt.xlabel("Longitud de uno de los lados")
plt.ylabel("Perímetro")
plt.legend(loc='best')
plt.show()
```
