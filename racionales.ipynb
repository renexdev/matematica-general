<h2>Funciones racionales</h2>


Otro tipo de funciones algebraicas son las funciones racionales. Una función racional es una función de la forma $r(x)=\frac{P(x)}{Q(x)}$, donde $P$ y $Q$ son polinomios. Asumimos que $P(x)$ y $Q(x)$ no tienen factor en común (al final del apunte se describirá esta sutileza que asumimos). 

Para que sea función, definimos el $Dom(r)$ como el conjunto de los $\mathbb{R}$ tales que no se anula el denominador, es decir cuando $Q(x)\neq 0$.


$\textbf{Ejemplo 1.}$ Encontar el dominio de las siguientes funciones racionales:

i)  $f(x)=\frac{1}{x}$

*Solución:* $Dom(f)=\{x \in \mathbb{R} / x \neq 0\}$

ii) $g(x)=\frac{2x^2+3}{x^2-1}$

*Solución:* $Dom(g)=\{x \in \mathbb{R} / x \neq 1 \land x \neq -1 \}$

iii)  $h(x)=\frac{x^3-2x^2+7}{x^2+1}$

*Solución:* $Dom(h)=\mathbb{R}$




<h3> Gráfica de funciones racionales</h3>

La gráfica de funciones racionales es diferente a las de las funciones polinomiales que vimos, a pesar de que se construyen a partir de éstas. Veamos como graficar la función $f(x)=\frac{1}{x}$ del ejemplo **3-i**. Como la función no está definida para $x=0$ nos va a interesar estudiar el comportamiento de la función en las cercanías donde el denominador se hace cero. Recordar la definición de entorno que utilizamos en el apunte anterior. En un curso de cálculo se estudiará la definición de entorno. Podemos entonces elaborar una tabla como la que se muestra a continuación para estudiar el comportamiento de $f(x)$ en un entorno alrededor del punto de interés, $x=0$. La tabla de la izquierda explora los valores que toma función cuando la variable $x$ se acerca a $0$ por el lado izquierdo de la recta de los reales, mientras que la tabla de la derecha lo hace cuando se acerca $x$ por la derecha.

El comportamiento observado puede describirse como sigue:

*Tabla 1)* $f(x) \rightarrow -\infty$ cuando $x\rightarrow 0^-$

y se lee la función tiende a menos infinito cuando $x$ se aproxima a 0 por la izquierda

*Tabla 2)* $f(x) \rightarrow \infty$ cuando $x\rightarrow 0^+$

y se lee la función tiende a infinito cuando $x$ se aproxima a 0 por la derecha

Veamos la gráfica de la función construída a partir de armar una tabla de pares ordenados $(x_0,f(x_0))$. 

```python

%matplotlib inline
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt

def racional(x):
    return 1/(x*1.)

plt.figure()
#Eje horizontal
plt.axhline(0, color='black')
#Eje vertical
plt.axvline(0, color='black')
#Grafico la funcion por partes
x1 = np.linspace(-6, -1e-8, 50)
x2 = np.linspace(1e-8, 6, 50)
plt.plot(x1,racional(x1), 'r', lw= 2, label='$f(x)=1/x$')
plt.plot(x2,racional(x2), 'r',lw= 2)
#Asíntota vertical
plt.axvline(0, linestyle='--',color='blue',label='asíntotal horizontal/vertical')
#Asíntota horizontal
plt.axhline(0, linestyle='--',color='blue')
#Establezco limites para graficar
plt.xlim(-6, 6)
plt.ylim(-5, 5)
#Titulo y nombre a los ejes
plt.xlabel("x")
plt.ylabel("y")
plt.legend(bbox_to_anchor=(1.01, 1), loc='upper left')
plt.show()
```

De la gráfica de $f(x)=\frac{1}{x}$ se ve que la función toma valores cada vez más grandes positivamente y negativamente en las cercanías del valor donde la función tiene problemas. Cuando se estudie la noción de límite de una funcion en un curso de cálculo se comprenderán mejor estos conceptos, nosotros nos limitaremos a dar una introducción intuitiva del comportamiento de la función.

En la gráfica puede observarse que dicho comportamiento puede describirse por una recta horizontal ficticia en $x=0$ 
a la que se aproxima continuamente la gráfica de la función; es decir que la distancia entre las dos tiende a ser cero (0), a medida que se extienden indefinidamente. Esta recta se conoce como **asíntota horizontal**. De forma análoga podemos definir intuitivamente una **asíntota vertical** y proponemos una recta $y$ a la que se aproxima continuamente la gráfica de la función.

Notar que si bien suelen representarse en un mismo sistema de coordenadas, las asíntotas NO forman parte de la expresión analítica de la función, por lo que -en numerosos ejemplos- no están incluidas explícitamente dentro de la gráfica, o bien se las indica con una línea punteada.

**Transformaciones lineales fraccionarias para graficar**. Una función racional de la forma $r(x)=\frac{ax+b}{cx+d}$ puede graficarse a partir de aplicar transformaciones a la función $f(x)=1/x$.. Veamos el siguiente ejemplo.

$\textbf{Ejemplo 2.}$ Graficar la función $r(x)=\frac{2}{x-3}$.

*Solución:* Tratemos de reescribir la función  $r(x)$ como función de $f(x)$.

$r(x)=\frac{2}{x-3}=2\frac{1}{x-3}=2f(x-3)$

De lo obtenido se puede ver que podemos graficar $r(x)$ corriendo la función $f$ en 3 unidades a la derecha y estirando el eje de las $y$s en un factor 2. 

```python

%matplotlib inline
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt

def racional(x):
    return 2/(x*1.-3.)

plt.figure()
#Eje horizontal
plt.axhline(0, color='black')
#Eje vertical
plt.axvline(0, color='black')
#Grafico la funcion por partes
x1 = np.linspace(-6,2.9999, 50)
x2 = np.linspace(3.0001, 6, 50)
plt.plot(x1,racional(x1), 'r', lw= 2, label='$f(x)=2/(x-3)$')
plt.plot(x2,racional(x2), 'r',lw= 2)

#Asíntota vertical
plt.axvline(3, linestyle='--',color='blue',label='asíntotal horizontal/vertical')
#Asíntota horizontal
plt.axhline(0, linestyle='--',color='blue')

#Establezco limites para graficar
plt.xlim(-2, 6)
plt.ylim(-5, 5)
#Titulo y nombre a los ejes
plt.title("Ejemplo 2")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(bbox_to_anchor=(1.01, 1), loc='upper left')
plt.show()
```


<h2> Cómo encontramos las asíntotas de las funciones racionales?</h2> 

Sea la función racional $r$ : $\mathbb{R} \rightarrow \mathbb{R}$ definida como $r(x)=\frac{a_nx^n+a_{n-1}x^{n-1}+...+a_1x+a_0}{b_mx^m+b_{m-1}x^{m-1}+...+b_1x+b_0}$, tenemos que,

1) Las asíntotas verticales de $r$ están en las líneas $x=x_i$ donde $\{x_i\}_1^N$ son $N$ ceros del denominador

2) Si $n<m$ entonces $r$ tiene una asíntota horizontal en $y=0$

3) Si $n=m$ entonces $r$ tiene una asíntota horizontal en $y=\frac{a_n}{b_n}$, siendo $a_n$ y $b_n$ los coeficientes dominantes de los polinomios del numerador y del denominador, respectivamente

4) Si $n>m$ entonces $r$ no tiene asíntota horizontal

$\textbf{Ejemplo 3.}$ Graficar y encontrar las asíntotas de la función $f(x)=\frac{3x^2-2x-1}{2x²+3x-2}$.

*Solución:* Primero, de ser posible, factorizamos el numerador y el denominador

$f(x)=\frac{3x^2-2x-1}{(2x-1)(x+2)}$

Por lo que la función $f$ tendrá dos asíntotas en $x=1/2$ y $x=-2$.

Para determinar si tiene asíntotas horizontales evaluamos el grado de los polinomios del numerador y del denominador, como tienen el mismo grado, tomamos el cociente de sus coeficientes dominantes, quedando así $y=3/2$.

Veamoslo en la gráfica de la función.

```python
%matplotlib inline
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt

def racional(x):
    return (3*x**2-2*x-1)/(2*x**2+3*x-2)

plt.figure()
#Eje horizontal
plt.axhline(0, color='black')
#Eje vertical
plt.axvline(0, color='black')
#Grafico la funcion por partes
x1 = np.linspace(-6,-2.00001, 50)
x2 = np.linspace(-1.99999, 0.49999999, 50)
x3 = np.linspace(0.5001, 6, 50)

plt.plot(x1,racional(x1), 'r', lw= 2, label='$f(x)=(3x^2-2x-1)/(2x^2+3x-2)$')
plt.plot(x2,racional(x2), 'r',lw= 2)
plt.plot(x3,racional(x3), 'r',lw= 2)

#Asíntota vertical en x=-2 y x=1
plt.axvline(0.5, linestyle='--',color='blue',label='asíntotal horizontal/vertical')
plt.axvline(-2, linestyle='--',color='blue')

#Asíntota horizontal en y=2
plt.axhline(3/2., linestyle='--',color='blue')

#Establezco limites para graficar
plt.xlim(-6, 6)
plt.ylim(-15, 20)
#Titulo y nombre a los ejes
plt.title("Ejemplo 3")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(bbox_to_anchor=(1.01, 1), loc='upper left')
plt.show()
```

<h2>Cómo graficamos las funciones racionales?</h2>

**Paso 1)** Factorizar el numerador y el denominador

**Paso 2)** Encontrar los ceros de la función y donde intecepta el eje de las $y$ (Evaluar $f(0)$)

**Paso 3)** Encontar las asíntotas verticales en los ceros del denominador. Evalue qué valores toma la función en las cercanías de esos puntos y ver si $y$ $\rightarrow \pm \infty$ en cada lado de la asíntota vertical.

**Paso 4)** Utilizar el método descrito anteriormente para determinar si la función tiene asíntotas horizontales.

**Paso 5)** Realize un bosquejo con lo obtenido de los pasos anteriores y agregue tantos puntos como sean necesarios para completar la gráfica de la función.

$\textbf{Ejemplo 3.}$ Grafique $f(x)=\frac{2x^2+7x-4}{x^2+x-2}$ y determine el dominio de la función.

*Solución:* Factorizando la función queda $f(x)=\frac{(2x-1)(x+4)}{(x-1)(x+2)}$.  
Los ceros de la función son en $x=1/2$ y $x=-4$, además $f(0)=2$.

Las asíntotas verticales ocurren en $x=-2$ y $x=1$ y el comportamiento de la función en entornos a esos puntos es el siguiente

La tabla muestra en la segunda fila, el signo del comportamiento que tendrá la función cuando el valor de la variable independiente se acerque a $x=-2$ y $x=1$ por derecha y por izquierda, esto se hace por ejemplo tomando un valor de prueba digamos -2.001 para $-2^-$ (qué valores le pondría a $-2^+$, $1^-$ y $1^+$?) y evaluar en todos los parentesis que signo queda, finalmente realizar el producto de signos, quedando negativo si hay un número impar de signos negativos y positivo en el caso par de signos negativos. La tercer fila muestra el valor al que tiende la función con los signos obtenidos anteriormente.

Como el grado del polinomio del numerador es igual al del denominador, tenemos que realizar el cociente de los coeficientes de los términos dominantes que da 2, por lo tanto la función tendrá una asíntota horizontal en $y=2$
La gráfica se muestra a continuación y el $Dom(f)=\{x \in \mathbb{R}/x\neq1 \land x\neq-2\}$

```python
%matplotlib inline
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt

def racional(x):
    return (2*x**2+7*x-4)/(x**2+x-2.)

plt.figure()
#Eje horizontal
plt.axhline(0, color='black')
#Eje vertical
plt.axvline(0, color='black')
#Grafico la funcion por partes
x1 = np.linspace(-10,-2.00001, 50)
x2 = np.linspace(-1.999999, 0.9999999, 50)
x3 = np.linspace(1.0001, 10, 50)

plt.plot(x1,racional(x1), 'r', lw= 2, label='$f(x)=(2x^2+7x-4)/(x^2+x-2)$')
plt.plot(x2,racional(x2), 'r',lw= 2)
plt.plot(x3,racional(x3), 'r',lw= 2)

#Asíntota vertical en x=-2 y x=1
plt.axvline(-2, linestyle='--',color='blue',label='asíntotal horizontal/vertical')
plt.axvline(1, linestyle='--',color='blue')

#Asíntota horizontal en y=2
plt.axhline(2, linestyle='--',color='blue')

#Establezco limites para graficar
plt.xlim(-10, 10)
plt.ylim(-5, 15)
#Titulo y nombre a los ejes
plt.title("Ejemplo 3")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(bbox_to_anchor=(1.01, 1), loc='upper left')
plt.show()
```

<h2>Algunos detalles a tener en cuenta ...</h2>

**Qué ocurre en los casos donde tenemos factores comunes en el numerador y en el denominador?.** Por definición no podremos evaluar la función donde no este definida y se deberá redefinir el dominio tal que no incluya a esos puntos. Por ejemplo la función $f(x)=\frac{x-2}{x^2-2x}$ no estará definida en $x=2$ ni en $x=0$ y la gráfica de la función presentará un hueco en  $x=2$ y una asíntota vertical en en $x=0$.

**Existe otro tipo de asíntotas?.** Sí y ocurren cuando las funciones racionales $r(x)=P(x)/Q(x)$ tienen el polinomio $P(x)$, del numerador, con un grado más que el del polinomio del denominador, $Q(x)$. El análisis de estos casos se tratará en un curso de cálculo cuando se estudie el concepto de límite. Por ejemplo, la función $f(x)=\frac{x^2-4x-5}{x-3}$ además de tener una asíntota vertical en $x=3$ tiene una asíntota oblicua descrita por $y=x-1$.
