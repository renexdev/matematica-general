<h2>Funciones con Dominio Real</h2>
<h3>Funciones racionales</h3>


Otro tipo de funciones algebraicas son las funciones racionales. Una función racional es una función de la forma $r(x)=\frac{P(x)}{Q(x)}$, donde $P$ y $Q$ son polinomios. Asumimos que $P(x)$ y $Q(x)$ no tienen factor en común. 

El $Dom(r)$ es el conjunto de los $\mathbb{R}$ tales que no se anula el denominador, es decir cuando $Q(x)\neq 0$.


$\textbf{Ejemplo 1.}$ Encontar el dominio de las siguientes funciones racionales:

i)  $f(x)=\frac{1}{x}$

*Solución:* $Dom(f)=\{x \in \mathbb{R} / x \neq 0\}$

ii) $g(x)=\frac{2x^2+3}{x^2-1}$

*Solución:* $Dom(g)=\{x \in \mathbb{R} / x \neq 1 \land x \neq -1 \}$

iii)  $h(x)=\frac{x^3-2x^2+7}{x^2+1}$

*Solución:* $Dom(h)=\mathbb{R}$




<h2> Gŕafica de funciones racionales</h2>

La gráfica de funciones racionales es diferente a las de las funciones polinomiales, a pesar de que se constuye a partir de éstas. Veamos como graficar la función $f(x)=\frac{1}{x}$ del ejemplo **3-i**. Nos va a interesar ver el comportamiento de la función en las cercanías donde el donominador se hace cero, es decir cerca de $x=0$. Podemos elaborar una tabla como la que se muestra a continuación para estudiar el comportamiento de $f(x)$ en un entorno a $x=0$. La tabla de izquierda explora los valores que toma funcón cuando la variale x se acerca a 0 por el lado izquierdo, mientras que la tabla de la derecha lo hace por el lado derecho.

El comportamiento observado puede describirse como sigue:

1) $f(x) \rightarrow -\infty$ cuando $x\rightarrow 0^-$

y se lee la función tiende a menos infinito cuando $x$ se aproxima a 0 por la izquierda

2) $f(x) \rightarrow \infty$ cuando $x\rightarrow 0^+$

y se lee la función tiende a infinito cuando $x$ se aproxima a 0 por la derecha

Veamos la gráfica de la función (si agrega un par de puntos para la variable $x$ y los evalúa también la puede construir en papel)

```python

%matplotlib inline
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt

def racional(x):
    return 1/(x*1.)

plt.figure()
#Eje horizontal
plt.axhline(0, color='black')
#Eje vertical
plt.axvline(0, color='black')
#Grafico la funcion por partes
x1 = np.linspace(-6, -1e-8, 50)
x2 = np.linspace(1e-8, 6, 50)
plt.plot(x1,racional(x1), 'r', lw= 2, label='$f(x)=1/x$')
plt.plot(x2,racional(x2), 'r',lw= 2)
#Asíntota vertical
plt.axvline(0, linestyle='--',color='blue',label='asíntotal horizontal/vertical')
#Asíntota horizontal
plt.axhline(0, linestyle='--',color='blue')
#Establezco limites para graficar
plt.xlim(-6, 6)
plt.ylim(-5, 5)
#Titulo y nombre a los ejes
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc='upper left')
plt.show()
```

De la gráfica de $f(x)=\frac{1}{x}$ se ve que la función toma valores cada vez más grandes o más chicos en las cercanías del valor donde la función tiene problemas (cuando se estudie la noción de límite de una funcion en un curso de cálculo se comprenderán mejor estos conceptos). Intuitivamente puede observarse que hay una recta vertical en $x=0$ que la función nunca toca (no puede intersectarla). Esta recta se conoce como **asíntota horizontal**.

De forma análoga podemos definir intuitivamente una **asíntota vertical** y proponemos una recta $y$ igual a una constante tal que la función no puede intersectarla.

**Transformaciones lineales fraccionarias para graficar**. Una función racional de la forma $r(x)=\frac{ax+b}{cx+d}$ puede graficarse a partir de aplicar transformaciones a la función $f(x)=1/x$.. Veamos el siguiente ejemplo.

$\textbf{Ejemplo 2.}$ Graficar la función $r(x)=\frac{2}{x-3}$.

*Solución:* Tratemos de reescribir la función  $r(x)$ como función de $f(x)$.

$r(x)=\frac{2}{x-3}=2\frac{1}{x-3}=2f(x-3)$

De lo obtenido se puede ver que podemos graficar $r(x)$ corriendo la función $f$ 3 unidades a la derecha y estirando el eje en un factor 2. 

```python

%matplotlib inline
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt

def racional(x):
    return 2/(x*1.-3.)

plt.figure()
#Eje horizontal
plt.axhline(0, color='black')
#Eje vertical
plt.axvline(0, color='black')
#Grafico la funcion por partes
x1 = np.linspace(-6,2.9999, 50)
x2 = np.linspace(3.0001, 6, 50)
plt.plot(x1,racional(x1), 'r', lw= 2, label='$f(x)=1/x$')
plt.plot(x2,racional(x2), 'r',lw= 2)

#Asíntota vertical
plt.axvline(3, linestyle='--',color='blue',label='asíntotal horizontal/vertical')
#Asíntota horizontal
plt.axhline(0, linestyle='--',color='blue')

#Establezco limites para graficar
plt.xlim(-2, 6)
plt.ylim(-5, 5)
#Titulo y nombre a los ejes
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc='best')
plt.show()
```


<h2> Cómo encontramos las asíntotas de las funciones racionales?</h2> 

Sea la función racional $r$ : $\mathbb{R} \rightarrow \mathbb{R}$ definida como $r(x)=\frac{a_nx^n+a_{n-1}x^{n-1}+...+a_1x+a_0}{b_nx^n+b_{n-1}x^{n-1}+...+b_1x+b_0}$, tenemos que,

1) Las asíntotas verticales de $r$ están en las líneas $x=x_i$ donde $\{x_i\}_1^N$ son $N$ ceros del denominador

2) Si $n<m$ entonces $r$ tiene una asíntota horizontal en $y=0$

3) Si $n=m$ entonces $r$ tiene una asíntota horizontal en $y=\frac{a_n}{b_n}$, siendo $a_n$ y $b_n$ los coeficientes dominantes de los polinomios del numerador y del denominador respectivamente

4) Si $n>m$ entonces $r$ no tiene una asíntota horizontal

$\textbf{Ejemplo 3.}$ Graficar y encontrar las asíntotas de la función $f(x)=\frac{3x^2-2x-1}{2x²+3x-2}$.

*Solución:* Primero, de ser posible, factorizamos el numerador y el denominador

$f(x)=\frac{3x^2-2x-1}{(2x-1)(x+2)}$

Por lo que la función $f$ tendrá dos asíntotas en $x=1/2$ y $x=2$.

Para determinar si tiene asíntotas horizontales evaluamos el grado de los polinomios del numerador y del denominaodr, como tienen el mismo grado, tomamos el cociente de sus coeficientes dominantes, quedando así $y=3/2$.

Veamoslo en la gráfica de la función.

```python

%matplotlib inline
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt

def racional(x):
    return 2/(x*1.-3.)

plt.figure()
#Eje horizontal
plt.axhline(0, color='black')
#Eje vertical
plt.axvline(0, color='black')
#Grafico la funcion por partes
x1 = np.linspace(-6,2.9999, 50)
x2 = np.linspace(3.0001, 6, 50)
plt.plot(x1,racional(x1), 'r', lw= 2, label='$f(x)=1/x$')
plt.plot(x2,racional(x2), 'r',lw= 2)

#Asíntota vertical
plt.axvline(3, linestyle='--',color='blue',label='asíntotal horizontal/vertical')
#Asíntota horizontal
plt.axhline(0, linestyle='--',color='blue')

#Establezco limites para graficar
plt.xlim(-2, 6)
plt.ylim(-5, 5)
#Titulo y nombre a los ejes
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc='best')
plt.show()
```
