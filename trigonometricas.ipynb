<h2>Funciones trigonométricas</h2>

Las funciones que no son algebraicas se denominan trascendentes. Reciben este nombre porque las operaciones que se efectúan sobre la variable independiente trascienden el campo del álgebra. Estudiaremos las funciones trigonométricas y sus inversas.

Consideremos un punto $P$ de la circunsferencia de radio unitario con centro en el origen de coordenadas.

Cuando $P$ toma diversas posiciones sobre la circunferencia, las coordenas de $x$ e $y$ varían entre -1 y 1. Una manera alternativa de determinar en forma unívoca la posición de $P$ es por el ángulo $\alpha$ que $\overline{OP}$ forma con el eje $x$ positivo. Consecuentemente las coordenas $x$ e $y$ también están determinadas por dicho ángulo.

Convenimos utilizar medidas de radianes para los ángulos. El conjunto de valores permitidos de $\alpha$ es el conjunto de los $\mathbb{R}$. Como se puede observar a cada valor de $\alpha$ le corresponde un único valor de $x$ y de $y$, por lo tanto podemos definir dos funciones con dominio en $\mathbb{R}$ e imagen en el rango $[-1,1]$: $\alpha \rightarrow x$ y $\alpha \rightarrow y$. A la primera función se la llama función **coseno** y es la proyección del segmento $\overline{OP}$ sobre el eje de las $x$s, y la segunda función es la función **seno** y es la proyección sobre el eje de las $y$s de dicho segmento.

**Algunas Propiedades**

* En el círculo unitario se observa que dada una posición del punto $P$ un incremento en $2\pi$ en la medida del ángulo deja sin cambio la posición del punto, y por lo tanto los valores de las coordenadas $x$ e $y$  no se alteran. Esto se escribe como,
$\cos{(\alpha+2k\pi)}=\cos{(\alpha)}$ y $\sin{(\alpha+2k\pi)}=\sin{(\alpha)}$ con $k \in \mathbb{Z}$ y  significa que las funciones seno y coseno son periódicas y el periodo es $2\pi$

* Si aplicamos ángulos con medidas $\alpha$ y $-\alpha$ radianes a las funciones tenemos: $\cos{(-\alpha)}=\cos{(\alpha)}$ y que el $\sin{(-\alpha)}=-\sin{(\alpha)}$. Lo que significa que la función coseno es una función par y la función seno es una función impar.

<center>

* Dado que $P(x,y)$ es un punto sobre el círculo unitario con centro en el origen de coordenadas, y la ecuación de la circunferencia es $x^2+y^2=1$, se tiene que $\cos^2{(\alpha)}+\sin^2{(\alpha)}=1$. Esta relación es fundamental en muchas relaciones que involucran las funciones $\sin(x)$ y $\cos(x)$ dado que se puede utilizar para cambiar de una función a la otra.

A continuación se muestra la gráfica de las funciones $sin(x)$ y $cos(x)$

```python
%matplotlib inline
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-4*np.pi,4*np.pi,0.1)   # start,stop,step

plt.figure()
#Eje horizontal
plt.axhline(0, color='black')
#Eje vertical
plt.axvline(0, color='black')
#Grafico la funcion por partes
x1 = np.linspace(-6, -1e-8, 50)
x2 = np.linspace(1e-8, 6, 50)
plt.plot(x/np.pi,np.sin(x), 'r', lw= 2, label='$f(x)=sin(x)$')
plt.plot(x/np.pi,np.cos(x), 'b',lw= 2, label='$f(x)=cos(x)$')
#Eje rango función
plt.axhline(1, linestyle= '--', color='black')
plt.axhline(-1, linestyle= '--', color='black')
#Establezco limites para graficar
plt.xlim(-2, 4)
plt.ylim(-1.1, 1.1)
#Titulo y nombre a los ejes
plt.xlabel("x (unidades de $\pi$)")
plt.ylabel("y")
plt.legend(bbox_to_anchor=(1.01, 1), loc='upper left')
plt.show()
```

Utilizando estas dos funciones es posible definir otras funciones trigonométricas que son útiles, a saber:

* La función tangente definda por $\tan{(\alpha)}=\sin{(\alpha)}/\cos{(\alpha)}$

* La función cotangente, $\cot{(\alpha)}=\cos{(\alpha)}/\sin{(\alpha)}$

* La función secante, $\sec{(\alpha)}=1/\cos{(\alpha)}$

* La función cosecante, $\csc{(\alpha)}=1/\sin{(\alpha)}$

Sus dominios son todos los números reales tales que no anulan el denominador, así:

$Dom(\tan)=Dom(\sec)=\{x \in \mathbb{R}/ x\neq \pm (2k+1)\pi/2, con\,k=0,1,2, ... \}$

$Dom(\cot)=Dom(\csc)=\{x \in \mathbb{R}/ x\neq \pm k\pi, con\,k=0,1,2, ... \}$

A continuación se muestra la gráfica de la función tangente. Notar que el rango de la función son los $\mathbb{R}$


```python
%matplotlib inline
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt

plt.figure()
#Eje horizontal
plt.axhline(0, color='black')
#Eje vertical
plt.axvline(0, color='black')
#Grafico la funcion por partes
delta = 1e-8
x1 = np.arange(-3*np.pi/2.+delta,-np.pi/2.-delta,0.1)   # start,stop,step
x2 = np.arange(-np.pi/2.+delta,np.pi/2.-delta,0.1)   # start,stop,step
x3 = np.arange(np.pi/2.+delta,3*np.pi/2.-delta,0.1)   # start,stop,step

plt.plot(x1/np.pi,np.tan(x1), 'r', lw= 2, label='$f(x)=tan(x)$')
plt.plot(x2/np.pi,np.tan(x2), 'r',lw= 2)
plt.plot(x3/np.pi,np.tan(x3), 'r',lw= 2)

#Asíntota vertical en x=-2 y x=1
plt.axvline(-1/2., linestyle='--',color='blue',label='asíntota vertical')
plt.axvline(1/2., linestyle='--',color='blue')
plt.axvline(3/2., linestyle='--',color='blue')
plt.axvline(-3/2., linestyle='--',color='blue')

#Establezco limites para graficar
plt.xlim(-5/np.pi, 5/np.pi)
plt.ylim(-15, 15)
#Titulo y nombre a los ejes
plt.title("Función tangente")
plt.xlabel("x (unidades de $\pi$)")
plt.ylabel("y")
plt.legend(bbox_to_anchor=(1.01, 1), loc='upper left')
plt.show()
```

$\textbf{Ejemplo 1.}$ Graficar la función $f(x)=\cos{(2x)}$, dar su dominio, rango y periodo.

*Solución:* El $Dom(f)= \mathbb{R}$ y su $Im(f)=\{x \in \mathbb{R}/ -1\leq x \leq  1\}$. Para determinar el periodo tenemos que recordar el periodo primitivo de $\cos{(x)}$ que es $2\pi$ queda $\cos{(2x)}=\cos{(2x+2[2\pi/2])}$, por lo que el periodo será $\pi$
Veamoslo en la gráfica de la función.

```python
%matplotlib inline
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-4*np.pi,4*np.pi,0.1)   # start,stop,step

plt.figure()
#Eje horizontal
plt.axhline(0, color='black')
#Eje vertical
plt.axvline(0, color='black')
#Grafico la funcion por partes
x1 = np.linspace(-6, -1e-8, 50)
x2 = np.linspace(1e-8, 6, 50)
plt.plot(x/np.pi,np.cos(x), 'k--', lw= 2, label='$f(x)=cos(x)$')
plt.plot(x/np.pi,np.cos(2*x), 'orange',lw= 2, label='$f(x)=cos(2x)$')
#Eje rango función
plt.axhline(1, linestyle= '--', color='black')
plt.axhline(-1, linestyle= '--', color='black')
#Establezco limites para graficar
plt.xlim(-1, 2)
plt.ylim(-1.1, 1.1)
#Titulo y nombre a los ejes
plt.xlabel("x (unidades de $\pi$)")
plt.ylabel("y")
plt.legend(bbox_to_anchor=(1.01, 1), loc='upper left')
plt.show()
```

<h2>Funciones trigonométricas inversas</h2>

Recordando lo estudiado sobre inyectividad de las funciones resulta fácil ver que las funciones trigonométricas no son inyectivas en sus dominios de definición, pero si los restringimos podemos encontrar intervalos donde éstas puedan resultar inyectivas y por lo tanto tener inversa.

Consideremos primeramente la función sno. Esta función es creciente en el intervalo $[-\pi/2,\pi/2]$ y por lo tanto inyectiva. Entonces si consideramos dicho intervalo para la función podremos definir su inversa que se denomina arcoseno, y se lo denota por $f^{-1}(x)=\arcsin{(x)}$ (o $\sin^{-1}{(x)}$, no confundir con la ley de potencias!).

Tenemos que $Dom(\arcsin)=[-1,1]$ y la $Im(\arcsin)=[-\pi/2,\pi/2]$. Observar que $y=\arcsin{(x)} \Leftrightarrow x=\sin{(x)}\,para\,y\in\,[-\pi/2,\pi/2]$. O bien, tenemos que $\arcsin{(\sin{y})}=y\,para\,y\in\,[-\pi/2,\pi/2]$ $\land$ $\sin{(\arcsin{x})}=x\,para\,x\in\,[-1,1]$


```python
%matplotlib inline
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt
x1 = np.arange(-np.pi/2.,np.pi/2.,0.1)   # start,stop,step
x2 = np.arange(-1.,1.,0.1)   # start,stop,step

plt.figure()
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.5, hspace=None)
#Grafico la funcion por partes
plt.subplot(2, 2, 1)
#Eje horizontal
plt.axhline(0, color='black')
#Eje vertical
plt.axvline(0, color='black')
plt.plot(x1/np.pi,np.sin(x1), 'b', lw= 2, label='$f(x)=sin(x)$')
plt.xlim(-0.5, 0.5)
plt.ylim(-1.1, 1.1)
plt.xlabel("x (unidades de $\pi$)")
plt.ylabel("y")
plt.legend(bbox_to_anchor=(2.51, 1.3), loc='upper left')

plt.subplot(2, 2, 2)
#Eje horizontal
plt.axhline(0, color='black')
#Eje vertical
plt.axvline(0, color='black')
plt.plot(x2,np.arcsin(x2)/np.pi, 'orange',lw= 2, label='$f(x)=arcsin(x)$')
plt.xlim(-1.5, 1.5)
plt.ylim(-0.5, .5)
plt.xlabel("x ")
plt.ylabel("y (unidades de $\pi$)")
#Eje rango función
#Establezco limites para graficar

#Titulo y nombre a los ejes

plt.legend(bbox_to_anchor=(1.01, 1.0), loc='upper left')
plt.show()
```

De la misma manera, restringiendo convenientemente el dominio, se define la función inversa del coseno llamada arcocoseno, $y=\arccos{(x)}$ cuyo estudio debe realizarlo el alumno (**hacer!**)

Para el caso de la función tangente, como esta es creciente en el intervalo $(-\pi/2,\pi/2)$ la función tendrá inversa en dicho intervalo y se denomina arco tangent, $y=\arctan{(x)}$ con $Dom(\arctan)=\mathbb{R}$ e $Im(\arctan)=(-\pi/2,\pi/2)$. Tenemos que $y=\arctan{(x)} \Leftrightarrow x=\tan{(y)}\,para\,y\in\,[-\pi/2,\pi/2]$.

La gráfica es:

```python
%matplotlib inline
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt
x1 = np.arange(-np.pi/2.,np.pi/2.,0.1)   # start,stop,step
x2 = np.arange(-10.,10.,0.1)   # start,stop,step

plt.figure()
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.5, hspace=None)
#Grafico la funcion por partes
plt.subplot(2, 2, 1)
#Eje horizontal
plt.axhline(0, color='black')
#Eje vertical
plt.axvline(0, color='black')
plt.plot(x1/np.pi,np.tan(x1), 'b', lw= 2, label='$f(x)=tan(x)$')
plt.axvline(-1/2., linestyle='--', color='blue')
plt.axvline(1/2., linestyle='--',color='blue')
plt.xlim(-0.6, 0.6)
plt.ylim(-15, 15)
plt.xlabel("x (unidades de $\pi$)")
plt.ylabel("y")
plt.legend(bbox_to_anchor=(2.51, 1.3), loc='upper left')

plt.subplot(2, 2, 2)
#Eje horizontal
plt.axhline(0, color='black')
#Eje vertical
plt.axvline(0, color='black')
plt.plot(x2,np.arctan(x2)/np.pi, 'orange',lw= 2, label='$f(x)=arctan(x)$')
plt.axhline(-1/2., linestyle='--', color='blue')
plt.axhline(1/2., linestyle='--',color='blue')
plt.xlim(-15, 15)
plt.ylim(-0.6, .6)
plt.xlabel("x ")
plt.ylabel("y (unidades de $\pi$)")
#Eje rango función
#Establezco limites para graficar

#Titulo y nombre a los ejes

plt.legend(bbox_to_anchor=(1.01, 1.0), loc='upper left')
plt.show()
```
